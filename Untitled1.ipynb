```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import PIL.Image as Image
from skimage import io
from google.colab.patches import cv2_imshow
cv2_imshow
def get_pixel(img,center,x,y):
  new_value=0
  try:
    if img[x][y]>=center:
      new_value=1
  except:
    pass
  return new_value

def lbp_calculated_pixel(img, x, y):
    center = img[x, y]
    val_ar = [
        get_pixel(img, center, x-1, y-1),
        get_pixel(img, center, x-1, y),
        get_pixel(img, center, x-1, y+1),
        get_pixel(img, center, x, y+1),
        get_pixel(img, center, x+1, y+1),
        get_pixel(img, center, x+1, y),
        get_pixel(img, center, x+1, y-1),
        get_pixel(img, center, x, y-1)
    ]
    # values to decimal
    power = [1, 2, 4, 8, 16, 32, 64, 128]
    val = 0
    for i in range(len(val_ar)):
        val += val_ar[i] * power[i]
    return val


path="/content/WhatsApp Image 2024-01-27 at 8.20.26 AM.jpeg"

img_brg = cv2.imread(path,1)

height ,width,_=img_brg.shape


# Open the image
img_gray = cv2.cvtColor(img_brg, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)

img_lbp = np.zeros((height, width), np.uint8)

for i in range(0, height):
    for j in range(0, width):
        img_lbp[i, j] = lbp_calculated_pixel(img_gray, i, j)

plt.imshow(img_brg)
plt.show()

plt.imshow(img_lbp, cmap ="gray")
plt.show()

print("LBP Program is finished")


def get_pixel(img,center,x,y):
  new_value=0
  try:
    if img[x][y]>=center:
      new_value=1
  except:
    pass
  return new_value

def lbp_calculated_pixel(img, x, y):
    center = img[x, y]
    val_ar = [
        get_pixel(img, center, x-1, y-1),
        get_pixel(img, center, x-1, y),
        get_pixel(img, center, x-1, y+1),
        get_pixel(img, center, x, y+1),
        get_pixel(img, center, x+1, y+1),
        get_pixel(img, center, x+1, y),
        get_pixel(img, center, x+1, y-1),
        get_pixel(img, center, x, y-1)
    ]
    # values to decimal
    power = [1, 2, 4, 8, 16, 32, 64, 128]
    val = 0
    for i in range(len(val_ar)):
        val += val_ar[i] * power[i]
    return val


path="/content/WhatsApp Image 2024-01-27 at 8.20.28 AM.jpeg"

img_brg = cv2.imread(path,1)

height ,width,_=img_brg.shape


# Open the image
img_gray = cv2.cvtColor(img_brg, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)

img_lbp = np.zeros((height, width), np.uint8)

for i in range(0, height):
    for j in range(0, width):
        img_lbp[i, j] = lbp_calculated_pixel(img_gray, i, j)

plt.imshow(img_brg)
plt.show()

plt.imshow(img_lbp, cmap ="gray")
plt.show()

print("LBP Program is finished")
```
